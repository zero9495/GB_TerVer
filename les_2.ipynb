```python
from math import factorial
import numpy as np

def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))

def bernulli(n,k,p):
    q = 1-p
    return combinations(n, k) * p**k * q**(n-k)

def puasson(n,m,p):
    _lambda = n*p
    return (np.e**(-_lambda))*(_lambda**m)/(factorial(m))
```
