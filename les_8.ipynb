```python
groups = [
    {'name':'Футболисты', 'values':[173, 175, 180, 178, 177, 185, 183, 182]},
    {'name':'Хоккеисты', 'values':[177, 179, 180, 188, 177, 172, 171, 184, 180]},
    {'name':'Штангисты', 'values':[172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170]}
]
groups
```

```python
k = len(groups)
print('k = {}'.format(k))


n_list = [len(group['values']) for group in groups]
print('n_list = {}'.format(n_list))


n = sum(n_list)
print('n = {}'.format(n))


y_mean_list = [
    sum(group['values']) / n_list[i]
    for i, group in enumerate(groups)
]
print('y_mean_list = {}'.format(y_mean_list))


y_mean = sum([
    sum(group['values'])
    for group in groups
]) / n
print('y_mean = {}'.format(y_mean))
```

```python
s2 = sum([
    (val - y_mean)**2
    for group in groups
    for val in group['values'] 
])
print('s2 = {}'.format(s2))


s2f = sum([
    n_list[i] * (y_mean_list[i] - y_mean)**2
    for i in range(k)
])
print('s2f = {}'.format(s2f))

s2ost = sum([
    (val - y_mean_list[i])**2
    for i in range(k)
    for val in groups[i]['values']
])
print('s2ost = {}'.format(s2ost))

print('(s2f + s2ost) - s2 = {}'.format((s2f + s2ost) - s2))
```

```python
sigma2f = s2f / (k-1)
print('sigma2f = {}'.format(sigma2f))

sigma2ost = s2ost / (n-k)
print('sigma2ost = {}'.format(sigma2ost))

FH = sigma2f / sigma2ost
print('FH = {}'.format(FH))

from scipy.stats import f
dfn = k-1
dfd = n-k
f_crit = f.ppf([0.95], dfn, dfd)
print('f_crit = {}'.format(f_crit))

pvalue = f.sf(FH, dfn, dfd)
print('pvalue = {}'.format(pvalue))

if FH > f_crit:
    print('Difference between groups is statisticaly significance')
    print('Различия среднего роста среди взрослых футболистов, хоккеистов и штангистов есть.')
```
