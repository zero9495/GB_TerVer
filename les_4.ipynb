## Exercise 4

```python
import scipy.stats

M = 173
sigma = 8
distr = scipy.stats.norm(173, 8)

print("а). больше 182 см")
p = 1 - distr.cdf(182)
print(' '*4 + str(p))

print("б). больше 190 см")
p = 1 - distr.cdf(190)
print(' '*4 + str(p))

print("в). от 166 см до 190 см")
p1 = distr.cdf(166)
p2 = distr.cdf(190)
print(' '*4 + str(p2-p1))

print("г). от 166 см до 182 см")
p1 = distr.cdf(166)
p2 = distr.cdf(182)
print(' '*4 + str(p2-p1))

print("д). от 158 см до 190 см")
p1 = distr.cdf(158)
p2 = distr.cdf(190)
print(' '*4 + str(p2-p1))

print("е). не выше 150 см или не ниже 190 см")
p1 = distr.cdf(150)
p2 = 1 - distr.cdf(190)
print(' '*4 + str(p1+p2))

print("ё). не выше 150 см или не ниже 198 см")
p1 = distr.cdf(150)
p2 = 1 - distr.cdf(198)
print(' '*4 + str(p1+p2))

print("ж). ниже 166 см.")
p = distr.cdf(166)
print(' '*4 + str(p))
```
