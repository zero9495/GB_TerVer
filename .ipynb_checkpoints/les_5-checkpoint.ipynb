## Exercise 1

```python
import numpy as np
from scipy import stats

sigma = 16
confidence_level = 0.95
alpha = 1 - confidence_level
mu = 80
n = 256

Z = stats.norm.ppf(confidence_level + alpha/2)

sem = sigma/np.sqrt(n)
delta = Z*sem
print('{} +- {}'.format(mu, delta))
print('{}, {}'.format(mu-delta, mu+delta))

stats.norm.interval(confidence_level, loc=mu, scale=sem)
```

## Exercise 2

```python
X = np.array([6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1])
n = len(X)
confidence_level = 0.95
alpha = 1 - confidence_level

x_mean = X.mean()
s = X.std(ddof=1)
t = stats.t.ppf(confidence_level + alpha/2, df=n-1)

delta = t*s/np.sqrt(n)
print('{} +- {}'.format(x_mean, delta))
print('{}, {}'.format(x_mean-delta, x_mean+delta))

stats.t.interval(confidence_level, loc=x_mean, scale=stats.sem(X), df=n-1)
```

## Exercise 3

```python
mu = 17
alpha = 0.05
confidence_level = 1 - alpha
n = 100
x_mean = 17.5
variance = 4
sigma = np.sqrt(variance)

print('H0: 17 = 17.5')
print('H1: 17 < 17.5')

z_score = stats.norm.ppf(confidence_level)
print('z_score = {}'.format(z_score))

sem = np.sqrt(n) / sigma
Z = (x_mean - mu) * sem
print('Z = {}'.format(Z))

if z_score < Z:
    print('reject H0 hypothesis')
```

## Exercise 4

```python
mu = 200
X = np.array([202, 203, 199, 197, 195, 201, 200, 204, 194, 190])
n = len(X)
confidence_level = 0.99
alpha = 1 - confidence_level

x_mean = X.mean()
print('x_mean = {}'.format(x_mean))

print('H0: 198.5 = 200')
print('H1: 198.5 < 200')

z_score = -stats.t.ppf(confidence_level + alpha/2, df=n-1)
print('z_score = {}'.format(z_score))

s = X.std(ddof=1)
sem = np.sqrt(n) / s
Z = (x_mean - mu) * sem
print('Z = {}'.format(Z))

if z_score < Z:
    print('allow H0 hypothesis')
```
