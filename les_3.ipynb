## Exercise 1

```python
salaries = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]
arr = np.array(salaries)
arr
```

```python
# Lenght
n = arr.size
n
```

```python
# Mean
arr_mean = arr.sum() / n
arr_mean
```

```python
# Sum of squared deviations from the mean
squred_dev_sum = ((arr - arr_mean)**2).sum()
```

```python
# Uncorrected sample standard deviation
uncor_std = squred_dev_sum / n
uncor_std
```

```python
# Unbiased sample variance
variance = np.sqrt(uncor_std)
variance
```

```python
# Corrected sample standard deviation
cor_std = squred_dev_sum / (n-1)
cor_std
```

```python
from math import factorial

def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
# AA Bbbb
p1_a = combinations(5,2)/combinations(8,2)
p1_b = combinations(5,1)*combinations(7,3)/combinations(12,4)
p1= p1_a * p1_b

# Aa BBbb
p2_a = combinations(5,1)*combinations(3,1)/combinations(8,2)
p2_b = combinations(5,2)*combinations(7,2)/combinations(12,4)
p2= p2_a * p2_b

# aa BBBb
p3_a = combinations(3,2)/combinations(8,2)
p3_b = combinations(5,3)*combinations(7,1)/combinations(12,4)
p3= p3_a * p3_b

p1+p2+p3
```
