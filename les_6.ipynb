## Exercise 1

```python
import numpy as np
import pandas as pd

zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])

cov = np.mean(zp*ks) - np.mean(zp) * np.mean(ks)
print('cov = {}'.format(cov))

cov = np.cov(zp,ks, ddof=0)[0,1]
print('np.cov = {}'.format(cov))

corr = cov / (zp.std() * ks.std())
print('corr = {}'.format(corr))

corr = np.corrcoef(zp, ks)[0,1]
print('np.corrcoef = {}'.format(corr))

df = pd.DataFrame({'zp':zp, 'ks':ks})
corr = df.corr().iloc[0,1]
print('df.corr = {}'.format(corr))
```

## Exercise 2

```python
from scipy import stats

X = np.array([131, 125, 115, 122, 131, 115, 107, 99, 125, 111])
confidence_level = 0.95
alpha = 1 - confidence_level
x_mean = X.mean()
n = len(X)

x_mean = X.mean()
s = X.std(ddof=1)
t = stats.t.ppf(confidence_level + alpha/2, df=n-1)

delta = t*s/np.sqrt(n)
print('{} +- {}'.format(x_mean, delta))
print('{}, {}'.format(x_mean-delta, x_mean+delta))

stats.t.interval(confidence_level, loc=x_mean, scale=stats.sem(X), df=n-1)
```

## Exercise 3

```python
variance = 25
sigma = np.sqrt(variance)
n = 27
mu = 174.2
confidence_level = 0.95
alpha = 1 - confidence_level

Z = stats.norm.ppf(confidence_level + alpha/2)

sem = sigma/np.sqrt(n)
delta = Z*sem
print('{} +- {}'.format(mu, delta))
print('{}, {}'.format(mu-delta, mu+delta))

stats.norm.interval(confidence_level, loc=mu, scale=sem)
```
